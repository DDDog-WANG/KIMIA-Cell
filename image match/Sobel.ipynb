```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("TitechIMG.jpg", 0)
x = cv2.Sobel(img,cv2.CV_16S,1,0)  #1,0代表只计算x方向计算边缘
y = cv2.Sobel(img,cv2.CV_16S,0,1)  #0,1代表只在y方向计算边缘
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
dst = cv2.addWeighted(absX,0.5,absY,0.5,0)

plt.figure(figsize=(10,10))
plt.figure(1)
plt.subplot(221)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(222)
plt.imshow(cv2.cvtColor(absX, cv2.COLOR_BGR2RGB))
plt.subplot(223)
plt.imshow(cv2.cvtColor(absY, cv2.COLOR_BGR2RGB))
plt.subplot(224)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
