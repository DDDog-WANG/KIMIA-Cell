# Compare each algorisms

* many methods to find key points of image
* compare each methods that calculate points' number, average distance

```python
import numpy as np
import cv2
from tqdm import tqdm
from matplotlib import pyplot as plt
import time,math,random
import pandas as pd 

import orb, sift, akaze, hashall, hist
```

```python
ORB_KnnMatch_matchpoints = []
ORB_KnnMatch_Averagedistance = []
ORB_match_matchpoints = []
ORB_match_Averagedistance = []

AKAZE_KnnMatch_matchpoints = []
AKAZE_KnnMatch_Averagedistance = []
AKAZE_match_matchpoints = []
AKAZE_match_Averagedistance = []

SIFT_KnnMatch_matchpoints = []
SIFT_KnnMatch_Averagedistance = []
SIFT_match_matchpoints = []
SIFT_match_Averagedistance = []

Hist_value = []

aHash_Hamming = []
pHash_Hamming = []
dHash_Hamming = []

index=[]
label=[]
```

```python
load="../Database/Else/PLAZA/"

for t in tqdm(range(1,10)):
    m=random.randint(1,10)
    n=random.randint(1,10)   
    if m==n: 
        label.append(1)
    else:
        label.append(0) #  label 
    
    mangle=random.randint(0,6)
    nangle=random.randint(0,6)
    img1="lungn"+str(m)+"_r"+str(mangle*30)
    img2="lungn"+str(n)+"_r"+str(nangle*30)    
    index.append(img1+" // "+img2) # index
    
    orb_li = orb.ORBexp(load,img1+".jpeg",img2+".jpeg") # ORB
    ORB_match_matchpoints.append(orb_li[0])
    ORB_match_Averagedistance.append(orb_li[1])
    ORB_KnnMatch_matchpoints.append(orb_li[2])
    ORB_KnnMatch_Averagedistance.append(orb_li[3])
    
    akaze_li = akaze.AKAZEexp(load,img1+".jpeg",img2+".jpeg") # AKAZE
    AKAZE_match_matchpoints.append(orb_li[0])
    AKAZE_match_Averagedistance.append(orb_li[1])
    AKAZE_KnnMatch_matchpoints.append(orb_li[2])
    AKAZE_KnnMatch_Averagedistance.append(orb_li[3])
    
    sift_li = sift.SIFTexp(load,img1+".jpeg",img2+".jpeg") # SIFT
    SIFT_match_matchpoints.append(orb_li[0])
    SIFT_match_Averagedistance.append(orb_li[1])
    SIFT_KnnMatch_matchpoints.append(orb_li[2])
    SIFT_KnnMatch_Averagedistance.append(orb_li[3])
     
    Hist_value.append(hist.Histexp(load,img1+".jpeg",img2+".jpeg")) # HIST
#     aHash_Hamming.append(hashall.aHashexp(load,img1+".jpeg",img2+".jpeg")) # HIST
```

```python
labels = ["pair","label",
          "ORB点数","ORB距離","ORBKnn点数","ORBKnn距離",
          "AKAZE点数","AKAZE距離","AKAZEKnn点数","AKAZEKnn距離",
          "SIFT点数","SIFT距離","SIFTKnn点数","SIFTKnn距離"]

array = np.array([index,label,ORB_match_matchpoints,ORB_match_Averagedistance,ORB_KnnMatch_matchpoints,ORB_KnnMatch_Averagedistance,
                  AKAZE_match_matchpoints,AKAZE_match_Averagedistance,AKAZE_KnnMatch_matchpoints,AKAZE_KnnMatch_Averagedistance,
                  SIFT_match_matchpoints,SIFT_match_Averagedistance,SIFT_KnnMatch_matchpoints,SIFT_KnnMatch_Averagedistance
                 ],dtype=object)
print(array.shape)
```

```python
df = pd.DataFrame(np.transpose(array), index=index, columns=labels)
df
```
